# **CSS453 Lab 5** - Laplace Transform

```python
import sympy as sp
sp.init_printing(use_latex=True)

def L(f):
  return sp.laplace_transform(f,t,s, noconds=True)

def invL(F):
  return sp.inverse_laplace_transform(F,s,t)

t,s = sp.symbols('t,s')
a = sp.symbols('a', real=True, positive=True)

def r(t):
  return t * sp.Heaviside(t)
```

  c. $\scriptsize ùë¶ = ùëí^{2ùë°}sin(2ùë°)$

```python
# 1.a.
y = t**2
f = L(y)
sp.plot(f, ylabel='f(t)')
f
```

```python
# 1.b.
y = sp.exp(-a*t) + sp.exp(-3*a*t)
L(y)
```

```python
# 1.d.
y = sp.exp(3*t) + sp.cos(6*t) - (sp.exp(-3*t)*sp.cos(6*t))
f = L(y)
sp.plot(f, ylabel='f(t)')
f
```

```python
# 1.e.
y = sp.Heaviside(t-2) + 2*sp.Heaviside(t-3) - 2*r(t-2)
f = L(y)
sp.plot(f, ylabel='f(t)')
f
```

Consider the two functions $\scriptsize ùëì(ùë°) = ùë¢(ùë°)ùë¢(3 ‚àí ùë°)$ and $\scriptsize ùëî(ùë°) = ùë¢(ùë°) ‚àí ùë¢(ùë° ‚àí 3)$.

#### **Question 6**

Solve the initial value problem using Laplace transform  
(a) $\scriptsize ùë¶' + 2ùë¶ = 4ùë°, ùë¶(0) = 3$  
(b) $\scriptsize ùë¶'' + 3ùë¶' + 2ùë¶ = 6ùëí^{‚àíùë°}, ùë¶(0) = 1, ùë¶'(0) = 2$  
(c) $\scriptsize ùë¶' + 4ùë¶ = ùëî(ùë°), ùë¶(0) = 2$, where $\scriptsize ùëî(ùë°) = 12$ if $\scriptsize(1 < ùë° < 3)$ else 0  

```python
y = sp.Function('y')(t)
y_ = sp.Derivative(y,t)     #  y_ == y'
y__ = sp.Derivative(y_,t)   # y__ == y''
```

```python
# 6.a.
f = y_ + 2*y - 4*t
L(f)
```

```python
# 6.b.
f = y__ + 3*y_ + 2*y - 6*sp.exp(-t)
L(f)
```

```python
# 6.c.
g = sp.Function('g')(t)
f = y_ + 4*y - g
L(f)
```
